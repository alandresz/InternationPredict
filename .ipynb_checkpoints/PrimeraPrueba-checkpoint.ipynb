```python
# Cargar datos
df_pacientes = pd.read_csv('Pacientes.csv', encoding='ISO-8859-1', sep=';')
```

```python
df_episodios = pd.read_csv('Episodios_diagnosticos.csv', encoding='ISO-8859-1', sep=';')
```

```python
df_signos = pd.read_csv('Signos_Vitales.csv', encoding='ISO-8859-1', sep=';')
```

```python
df_estudios = pd.read_csv('Estudios_complementarios.csv', encoding='ISO-8859-1', sep=';')
```

```python
# Explorar datos
print("Pacientes.csv:")
print(df_pacientes.head())
print("\nEpisodios_diagnosticos.csv:")
print(df_episodios.head())
print("\nSignos_Vitales.csv:")
print(df_signos.head())
print("\nEstudios_complementarios.csv:")
print(df_estudios.head())
```

```python
# --- Limpiar Episodios_diagnosticos ---
# Eliminar registros con TIPO_EPISODIO = '*'
df_episodios = df_episodios[df_episodios['TIPO_EPISODIO'] != '*']
print(df_episodios.head())
```

```python

# Crear variable objetivo (CLASE)
df_episodios['CLASE'] = df_episodios['TIPO_EPISODIO'].apply(lambda x: 1 if x == 'H' else 0)

# Reducir a nivel de paciente (si alguna vez se internÃ³, CLASE=1)
df_target = df_episodios.groupby('PACIENTE')['CLASE'].max().reset_index()

# --- Limpiar Estudios_complementarios ---
# Descartar ID_ITEM como indica el enunciado
df_estudios = df_estudios.drop(columns=['ID_ITEM'])
```

```python
# Opcional: Filtrar prÃ¡cticas frecuentes
frequent_practices = df_estudios['PRACTICA'].value_counts().head(20).index
df_estudios = df_estudios[df_estudios['PRACTICA'].isin(frequent_practices)]

# --- Limpiar Signos_Vitales ---
# Opcional: Filtrar descripciones frecuentes
frequent_signs=df_signos['DESCRIPCION'].value_counts().head(20).index
df_signos = df_signos[df_signos['DESCRIPCION'].isin(frequent_signs)]
```

```python
print(frequent_practices)
print(frequent_signs)
```

```python
print("df_episodios:")
print(df_episodios.head())
print(df_episodios.dtypes)
```

```python
print("df_target:")
print(df_pacientes.head())
print(df_pacientes.dtypes)
```

```python
print("df_target:")
print(df_target.head())
print(df_target.dtypes)
```

```python
# --- Generar dataset final ---
# Combinar datos de pacientes con target
df_final = pd.merge(df_pacientes, df_target, on='PACIENTE', how='inner')
```

```python
print("df_final:")
print(df_final.head())
print(df_final.dtypes)
```

```python
print(df_target.shape)
```

```python
print(df_signos['VALOR_RESULTADO'].dtype)
print(df_signos['VALOR_RESULTADO'].unique())
print(df_estudios['RESULTADO_NUMERICO'].dtype)
print(df_estudios['RESULTADO_NUMERICO'].unique())

print(df_signos.dtypes)
print(df_signos.shape)
print(df_estudios.dtypes)
print(df_estudios.shape)
```

```python
df_signos['VALOR_RESULTADO'] = pd.to_numeric(df_signos['VALOR_RESULTADO'], errors='coerce')
df_estudios['RESULTADO_NUMERICO'] = pd.to_numeric(df_estudios['RESULTADO_NUMERICO'], errors='coerce')
```

```python
print(df_signos['VALOR_RESULTADO'].isnull().sum())
print(df_estudios['RESULTADO_NUMERICO'].isnull().sum())
```

```python
# Agregar datos de signos vitales y estudios complementarios
# (Ejemplo: Promedios por paciente)
signos_agg = df_signos.groupby('PACIENTE').agg({'VALOR_RESULTADO': 'mean'}).rename(columns={'VALOR_RESULTADO': 'PROMEDIO_SIGNOS'})
df_final = pd.merge(df_final, signos_agg, on='PACIENTE', how='left')

estudios_agg = df_estudios.groupby('PACIENTE').agg({'RESULTADO_NUMERICO': 'mean'}).rename(columns={'RESULTADO_NUMERICO': 'PROMEDIO_ESTUDIOS'})
df_final = pd.merge(df_final, estudios_agg, on='PACIENTE', how='left')
```

```python
print(signos_agg.dtypes)
print(signos_agg.size)

print(estudios_agg.dtypes)
print(estudios_agg.size)
```

```python
# --- Modelado ---
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Separar variables predictoras y objetivo
X = df_final.drop(columns=['PACIENTE', 'CLASE', 'DOMICILIO', 'ALTURA', 'LOCALIDAD', 'PROVINCIA'])
y = df_final['CLASE']
```

```python
# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.dtypes)
```

```python
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)
```

```python
print(X_train.dtypes)
```

```python
print(X_test.isnull().sum())
print(X_test.isnull().sum())
```

```python
X_train.fillna(X_train.mean(), inplace=True)
X_test.fillna(X_test.mean(), inplace=True)
```

```python
# Entrenar modelo
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Evaluar modelo
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy del modelo: {accuracy}")
```

```python
# Guardar resultados
df_final[['PACIENTE', 'CLASE']].to_csv('Resultados_Clasificador.csv', index=False)
```

```python
# Asegurarnos de que PACIENTE sea la clave Ãºnica
df_pacientes = df_pacientes.drop_duplicates(subset='PACIENTE')

# Crear el dataframe final con EDAD y SEXO
df_final = df_pacientes[['PACIENTE', 'EDAD', 'SEXO']]

# Codificar SEXO (F -> 0, M -> 1)
df_final['SEXO'] = df_final['SEXO'].map({'F': 0, 'M': 1})

# Agregar cantidad de internaciones por paciente
internaciones_agg = df_episodios[df_episodios['TIPO_EPISODIO'] == 'H'].groupby('PACIENTE').size()
df_final = pd.merge(df_final, internaciones_agg.rename('CANTIDAD_INTERNACIONES'), on='PACIENTE', how='left')

# Agregar cantidad de episodios por paciente
episodios_agg = df_episodios.groupby('PACIENTE').size()
df_final = pd.merge(df_final, episodios_agg.rename('CANTIDAD_EPISODIOS'), on='PACIENTE', how='left')

# Agregar cantidad de estudios por paciente
estudios_agg = df_estudios.groupby('PACIENTE').size()
df_final = pd.merge(df_final, estudios_agg.rename('CANTIDAD_ESTUDIOS'), on='PACIENTE', how='left')

# Agregar cantidad de signos vitales por paciente
signos_agg = df_signos.groupby('PACIENTE').size()
df_final = pd.merge(df_final, signos_agg.rename('CANTIDAD_SIGNOS_VITALES'), on='PACIENTE', how='left')

# Rellenar valores faltantes con 0 (pacientes que no tienen internaciones, estudios, etc.)
df_final.fillna(0, inplace=True)
```

```python
print(df_final.dtypes)
print(df_final.head(10))
```

```python
df_final['TIPO_EPISODIO'] = (df_final['CANTIDAD_INTERNACIONES'] > 0).astype(int)
```

```python
df_final = df_final.drop(columns=['CANTIDAD_INTERNACIONES'])
```

```python
print(df_final.dtypes)
print(df_final.head(10))
```

```python
X = df_final[['EDAD', 'SEXO', 'CANTIDAD_EPISODIOS', 'CANTIDAD_ESTUDIOS', 'CANTIDAD_SIGNOS_VITALES']]
y = df_final['TIPO_EPISODIO']

```

```python
X = pd.get_dummies(X, columns=['SEXO'], prefix='SEXO')
```

```python
print(X.dtypes)
```

```python
from sklearn.model_selection import train_test_split

# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)

# Evaluar el modelo
accuracy = model.score(X_test, y_test)
print(f"Exactitud del modelo: {accuracy:.2f}")
```

```python
cv_score=cross_val_score(model, X_train, y_train, cv=10)
print(f"CV average score: {cv_score.mean():.2f}")
```

```python
y_pred=model.predict(X_test)
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
cr=classification_report(y_test, y_pred)
print(cr)
```
